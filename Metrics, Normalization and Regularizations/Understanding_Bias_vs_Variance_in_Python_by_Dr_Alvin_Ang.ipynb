#Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
```

#Step 2: Load the Boston Housing Data.csv

```python
  # Load the Boston housing dataset.
df = pd.read_csv('https://www.alvinang.sg/s/boston_housing_data.csv')
df
```

#Step 3: Train Test Split

```python
#X_train will sample 70% of the df
X_train = df.sample(frac=0.7)

#then, we drop off the X_train from the df (since its a subset of it)
#this is done by matching the similar index column rows via the .index
#then, we keep the remaining and store it into X_test

X_test = df.drop(X_train.index)
```

Screenshot_1.jpg

```python
y_train = X_train.pop('MEDV')

y_test = X_test.pop('MEDV')
```

#Step 4: Create the LR Model using SKLearn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

#Step 5: Train / Fit the model using Linear Regression (using X_train and y_train)

```python
# Train the model on the training data.
model.fit(X_train, y_train)
```

#Step 6: Use the Model on the X_test to Predict the y_pred

```python
# Predict the values of y using the model.
y_pred = model.predict(X_test)
```

#Step 7: Find the MSE

```python
import sklearn.metrics

mse = sklearn.metrics.mean_squared_error(y_test, y_pred)

print("The mean squared error is:", mse)
```
