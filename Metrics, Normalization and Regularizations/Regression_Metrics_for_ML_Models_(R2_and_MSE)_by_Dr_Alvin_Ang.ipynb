- MSE measures how close the predicted values are to the true values (doesn't involve Curve Fitting)
- R2 score is a measure of how well the predicted values fit the overall trend of the data. (Goodness-of-Fit) (involves Curve Fitting)
- The best metric to use depends on the specific application.



---



#1) Mean Squared Error (MSE)

- MSE of 0 indicates a perfect fit
- MSE of infinity indicates a completely inaccurate fit.
- However, these are extreme cases and are rarely seen in practice.

- A more realistic MSE range for regression models is between 0 and 100.
- MSE of 10 or less is considered to be a good fit
- MSE of 100 or more is considered to be a poor fit.

- Good or Bad depends on specific application.

- A larger data set will typically produce a lower MSE.
- A model with more features will typically produce a higher MSE.
- A more complex model will typically produce a higher MSE.

##1a) Just getting the MSE....

```python
from sklearn.metrics import mean_squared_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mean_squared_error(y_true, y_pred)

#0.375 is a very low error! which means great prediction!
```

##1b) If we want to plot to check the Error.... (OPTIONAL)

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot the data as a scatter plot
plt.scatter(y_true, y_pred, color='blue')

# Create the line
line = np.arange(-1, 9)
plt.plot(line, line, color='red')

# Add a title
plt.title('Scatter Plot with Line')

# Add labels to the axes
plt.xlabel('y_True Values')
plt.ylabel('y_Predicted Values')

# Show the plot
plt.show()

#anything on the red line indicates very good prediction
```

#2) R2

The R2 score is a number between 0 and 1
- 1 indicates a perfect fit
- 0 indicates no fit.

##2a) Just getting the R2...

```python
from sklearn.metrics import r2_score

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

r2_score(y_true, y_pred)

# 95% fitting! very good!
```

##2b) If we want to plot to check the R2....

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Generate some random X values
X = np.random.randint(-10, 10, 10)

# Reshape the X array from row to column
X = X.reshape(-1, 1)

# Generate some random y values
y = np.random.randint(-10, 10, 10)

# Reshape the y array from row to column
y = y.reshape(-1, 1)

# Create the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Plot the data
plt.scatter(X, y)
plt.plot(X, model.predict(X), color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Curve Fitting")
plt.show()

# Calculate the R2 score
r2 = r2_score(y, model.predict(X))
print("R2 score:", r2)
```
