#1) Using Scikit Learn Train Test Split

##Step 1: Bring in the Dataset

```python
import pandas as pd
X = pd.read_csv('https://www.alvinang.sg/s/boston_housing_data.csv')
X
```

##Step 2: Pop out the y

```python
y = X.pop('MEDV')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```
